**Play with initial conditions to determine how the Potential Temperature Profile is responding!**

[ a ] = [$m^{1/4}·s^{1/4}$] 

[ M$_{RL}$ ] = [$m/s$] 

[ $Q_{AK}$ ] = [$K·m$] 

[ t ] = [$s$] 

```python
from numpy.lib.index_tricks import MGridClass
import numpy as np
import matplotlib.pyplot as plt


z=np.linspace(0,800,800)
def SPTP(a,Mrl,t,Qak):
  He=a*(Mrl)**(3/4)*t**(1/2)
  ΔΘs=Qak/He
  ΔΘ=ΔΘs*np.exp(-z/He)
  print('For wind=%.3f and Qak=%.3f' %(Mrl,Qak))
  print('He=%.3f' %(He))
  print('ΔΘs=%.3f' %(ΔΘs))
  plt.plot(ΔΘ,z,label='For wind=%.3f and Qak=%.3f' %(Mrl,Qak))
  plt.legend()


# Give the initial initial conditions SPTP(a,Mrl,t,Qak)
SPTP(0.15,10,43200,-1000)
SPTP(0.15,5,43200,-1000)
SPTP(0.15,3,43200,-1000)
```
