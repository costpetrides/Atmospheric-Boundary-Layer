```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/Initialvalue.csv'
df = pd.read_csv(url)
print(df)
```

```python
plt.figure(figsize=(13,2))

plt.xlabel('Hight [m]')
plt.ylabel('R.C.S [arb. unit]')

plt.plot(df['Z'], df['N1'], 'b', label='Normalized signal - 355 nm')
plt.plot(df['Z'], df['M1'], 'r', label='Molecular signal - 355 nm')

plt.xlim(0,30000)
plt.legend()

plt.show()
```
