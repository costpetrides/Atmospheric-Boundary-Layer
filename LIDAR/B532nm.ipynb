```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/B532.csv'
df = pd.read_csv(url)
print(df)   
```

# Backscatter Coefficient.

```python
plt.figure(figsize=(4,4))

plt.xlabel('Backscatter Coeff. [$Mm^{-1} sr^{-1}$]')
plt.ylabel('Hight [m]')

plt.plot(df['LR120'], df['Z'], 'purple', label='LR=120')
plt.plot(df['LR100'], df['Z'], 'c', label='LR=100')
plt.plot(df['LR80'], df['Z'], 'b', label='LR=80')
plt.plot(df['LR60'], df['Z'], 'g', label='LR=60')
plt.plot(df['LR40'], df['Z'], 'r', label='LR=40')
plt.plot(df['LR20'], df['Z'], 'black', label='LR=20')

plt.ylim(0,5000)
plt.xlim(0,4.5e-6)
plt.legend()

plt.show()
```

# Extinction Coefficient

```python
plt.figure(figsize=(4,4))

plt.xlabel('Extinction Coeff. [$Mm^{-1}$]')
plt.ylabel('Hight [m]')

plt.plot(df['LR120']*120, df['Z'], 'purple', label='LR=120')
plt.plot(df['LR100']*100, df['Z'], 'c', label='LR=100')
plt.plot(df['LR80']*80, df['Z'], 'b', label='LR=80')
plt.plot(df['LR60']*60, df['Z'], 'g', label='LR=60')
plt.plot(df['LR40']*40, df['Z'], 'r', label='LR=40')
plt.plot(df['LR20']*20, df['Z'], 'black', label='LR=20')

plt.ylim(0,5000)
plt.xlim(0,2.5e-4)
plt.legend()

plt.show()
```
