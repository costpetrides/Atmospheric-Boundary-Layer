```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/B355.csv'
df = pd.read_csv(url)
print(df)   
```

```python
plt.figure(figsize=(5,3))

plt.xlabel('Backscatter Coefficient [$Mm^{-1} sr^{-1}$]')
plt.ylabel('Hight [m]')

plt.plot(df['LR60'], df['Z'], 'g', label='LR=60')
plt.plot(df['LR40'], df['Z'], 'r', label='LR=40')
plt.plot(df['LR20'], df['Z'], 'black', label='LR=20')

plt.ylim(0,5000)
plt.legend()

plt.show()
```
