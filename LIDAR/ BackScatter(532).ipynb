```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/B532.csv'
df = pd.read_csv(url)
print(df)   
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Backscatter Coeff. [$Mm^{-1} sr^{-1}$]')
plt.ylabel('Hight [m]')

plt.plot(df['LR120'][0:50], df['Z'][0:50], 'purple', linestyle = ':' , alpha=0.5)   
plt.plot(df['LR120'][50:3993], df['Z'][50:3993], 'purple', label='LR=120') 

plt.plot(df['LR100'][0:50], df['Z'][0:50], 'c', linestyle = ':', alpha=0.5)
plt.plot(df['LR100'][50:3993], df['Z'][50:3993], 'c', label='LR=100')

plt.plot(df['LR80'][0:50], df['Z'][0:50], 'b', linestyle = ':', alpha=0.5)
plt.plot(df['LR80'][50:3993], df['Z'][50:3993], 'b', label='LR=80')

plt.plot(df['LR60'][0:50], df['Z'][0:50], 'g', linestyle = ':', alpha=0.5)
plt.plot(df['LR60'][50:3993], df['Z'][50:3993], 'g', label='LR=60')

plt.plot(df['LR40'][0:50], df['Z'][0:50], 'r', linestyle = ':', alpha=0.5)
plt.plot(df['LR40'][50:3993], df['Z'][50:3993], 'r', label='LR=40')

plt.plot(df['LR20'][0:50], df['Z'][0:50], 'black', linestyle = ':', alpha=0.5)
plt.plot(df['LR20'][50:3993], df['Z'][50:3993], 'black', label='LR=20')

plt.ylim(0,5000)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Relative Change %')
plt.ylabel('Hight [m]')

c120 = (df['LR120'][50:3993] - df['LR60'][50:3993]) / df['LR60'][50:3993]
c100 = (df['LR100'][50:3993] - df['LR60'][50:3993]) / df['LR60'][50:3993]
c80 = (df['LR80'][50:3993] - df['LR60'][50:3993]) / df['LR60'][50:3993]
c60 = (df['LR60'][50:3993] - df['LR60'][50:3993]) / df['LR60'][50:3993]
c40 = (df['LR40'][50:3993] - df['LR60'][50:3993]) / df['LR60'][50:3993]
c20 = (df['LR20'][50:3993] - df['LR60'][50:3993]) / df['LR60'][50:3993]


plt.plot(c120*100, df['Z'][50:3993], 'purple', label='LR=120') 

plt.plot(c100*100, df['Z'][50:3993], 'c', label='LR=100')

plt.plot(c80*100, df['Z'][50:3993], 'b', label='LR=80')

plt.plot(c60*100, df['Z'][50:3993], 'g', label='LR=60')

plt.plot(c40*100, df['Z'][50:3993], 'r', label='LR=40')

plt.plot(c20*100, df['Z'][50:3993], 'black', label='LR=20')

plt.ylim(0,6000)
plt.xlim(-50,50)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Absolute Change %')
plt.ylabel('Hight [m]')

c120 = abs(df['LR120'][50:3993] - df['LR60'][50:3993]) 
c100 = abs(df['LR100'][50:3993] - df['LR60'][50:3993]) 
c80 = abs(df['LR80'][50:3993] - df['LR60'][50:3993]) 
c60 = abs(df['LR60'][50:3993] - df['LR60'][50:3993]) 
c40 = abs(df['LR40'][50:3993] - df['LR60'][50:3993]) 
c20 = abs(df['LR20'][50:3993] - df['LR60'][50:3993]) 

plt.plot(c120, df['Z'][50:3993], 'purple', label='LR=120') 

plt.plot(c100, df['Z'][50:3993], 'c', label='LR=100')

plt.plot(c80, df['Z'][50:3993], 'b', label='LR=80')

plt.plot(c60, df['Z'][50:3993], 'g', label='LR=60')

plt.plot(c40, df['Z'][50:3993], 'r', label='LR=40')

plt.plot(c20, df['Z'][50:3993], 'black', label='LR=20')

plt.ylim(0,6000)
plt.legend()

plt.show()
```
