```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))
LR=[20,40,60,80,100,120]
O355=[0.17, 0.25, 0.3, 0.32, 0.34, 0.35]
O532=[0.07, 0.12, 0.17, 0.21, 0.24, 0.26]
O1664=[0.009, 0.02, 0.03, 0.04, 0.04, 0.05]

plt.scatter(LR,O355,label='355nm')
plt.scatter(LR,O532,label='532nm')
plt.scatter(LR,O1664,label='1064nm')
plt.xlabel('LR')
plt.ylabel('AOD')
plt.legend()
plt.show()
```
