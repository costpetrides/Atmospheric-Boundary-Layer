```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/E1064.csv'
df = pd.read_csv(url)
print(df) 
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Backscatter Coeff. [$Mm^{-1} sr^{-1}$]')
plt.ylabel('Hight [m]')

plt.plot(df['LR60'][50:3993], df['Z'][50:3993], 'g', label='LR=60')

plt.plot(df['LR40'][50:3993], df['Z'][50:3993], 'r', label='LR=40')

plt.plot(df['LR20'][50:3993], df['Z'][50:3993], 'black', label='LR=20')

plt.ylim(0,5000)
plt.xlim(0,0.00002)
plt.legend()
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Relative Change %')
plt.ylabel('Hight [m]')

c60 = (df['LR60'][85:3993] - df['LR60'][85:3993]) / df['LR60'][85:3993]
c40 = (df['LR40'][85:3993] - df['LR60'][85:3993]) / df['LR60'][85:3993]
c20 = (df['LR20'][85:3993] - df['LR60'][85:3993]) / df['LR60'][85:3993]

plt.plot(c60*100, df['Z'][85:3993], 'g', label='LR=60')

plt.plot(c40*100, df['Z'][85:3993], 'r', label='LR=40')

plt.plot(c20*100, df['Z'][85:3993], 'black', label='LR=20')

plt.ylim(0,4500)
plt.xlim(-120,120)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Absolute Change [$Mm^{-1} sr^{-1}$]')
plt.ylabel('Hight [m]')

c120 = abs(df['LR120'][85:3993] - df['LR60'][85:3993]) 
c100 = abs(df['LR100'][85:3993] - df['LR60'][85:3993]) 
c80 = abs(df['LR80'][85:3993] - df['LR60'][85:3993]) 
c60 = abs(df['LR60'][85:3993] - df['LR60'][85:3993]) 
c40 = abs(df['LR40'][85:3993] - df['LR60'][85:3993]) 
c20 = abs(df['LR20'][85:3993] - df['LR60'][85:3993]) 


plt.plot(c60, df['Z'][85:3993], 'g', label='LR=60')

plt.plot(c40, df['Z'][85:3993], 'r', label='LR=40')

plt.plot(c20, df['Z'][85:3993], 'black', label='LR=20')

plt.ylim(0,4500)
plt.legend()

plt.show()
```
