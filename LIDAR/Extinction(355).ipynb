```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/B355.csv'
df = pd.read_csv(url)
print(df)  
```

# Extinction Coefficient

```python
plt.figure(figsize=(4,4))

plt.xlabel('Extinction Coeff. [$Mm^{-1}$]')
plt.ylabel('Hight [m]')

plt.plot(df['LR60'][0:85]*60, df['Z'][0:85], 'g', linestyle = ':', alpha=0.5)
plt.plot(df['LR60'][85:3993]*60, df['Z'][85:3993], 'g', label='LR=60',linestyle = '-')

plt.plot(df['LR40'][0:85]*40, df['Z'][0:85], 'r', linestyle = ':', alpha=0.5)
plt.plot(df['LR40'][85:3993]*40, df['Z'][85:3993], 'r', label='LR=40',linestyle = '-')

plt.plot(df['LR20'][0:85]*20, df['Z'][0:85], 'black', linestyle = ':', alpha=0.5)
plt.plot(df['LR20'][85:3993]*20, df['Z'][85:3993], 'black', label='LR=20',linestyle = '-')

plt.ylim(0,5500)
plt.xlim(0,2e-4)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.xlabel('Relative Change %')
plt.ylabel('Hight [m]')

c60 = (df['LR60'][85:3993]*60 - df['LR60'][85:3993]*60) / df['LR60'][85:3993]*60
c40 = (df['LR40'][85:3993]*40 - df['LR60'][85:3993]*40) / df['LR60'][85:3993]*40
c20 = (df['LR20'][85:3993]*20 - df['LR60'][85:3993]*20) / df['LR60'][85:3993]*20

plt.plot(c60*100, df['Z'][85:3993], 'g', label='LR=60')

plt.plot(c40*100, df['Z'][85:3993], 'r', label='LR=40')

plt.plot(c20*100, df['Z'][85:3993], 'black', label='LR=20')

plt.ylim(0,5000)
plt.xlim(-120,120)
plt.legend()

plt.show()
```
