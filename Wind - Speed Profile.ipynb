# Wind - Speed Profile 

## Neutral Surface Layer (SL)

Screenshot 2023-01-25 at 3.24.41 PM.png

[ u$_{*}$ ]=[$m/s$] : friction velocity

[ k ]=0.4 : Von Karman constant 


[ z$_{o}$ ]=[$m$] : Roughness - length

```python
import numpy as np
import matplotlib.pyplot as plt

z=np.linspace(0.1,100,100)

def Uneutral(ustar,zo):
  k=0.4
  u=(ustar/k)*np.log(z/zo)
  print('For u*=%.3f and zo=%.3f' %(ustar,zo))
  plt.plot(u,z,label='For u*=%.3f and zo=%.3f' %(ustar,zo))
  plt.legend()

# Give the initial  conditions Uneutral(ustar,zo)
Uneutral(2.67,0.5)
Uneutral(2.67,0.2)
Uneutral(3,0.5)
```
