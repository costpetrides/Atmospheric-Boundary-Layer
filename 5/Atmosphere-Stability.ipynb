# Anisotropic and Isotropic Atmosphere

## 1.Mean value and Standard deviation of velocity.

```python
import numpy as np 
import matplotlib.pyplot as plt
          

t=np.linspace(0.1,1,10)                #Give the time matrix
v=np.array([2,-1,1,1,-3,-2,0,2,-1,1])  #Give the velocity matrix

μ=np.mean(v)
σv=np.std(v)

print('Mean value of v:  μ= %.3f m/s' %(μ))
print('Standard deviation of v: σv= %.3f m/s' %(σv))


vmin=min(v)
vmax=max(v)
N=len(v)



#Plot
plt.plot(t,v)
plt.title('velocity')
plt.xlabel('t [s]')
plt.ylabel('v [m/s]') 
plt.axhline(y=μ,color='red',label='mean value')
plt.legend() 
plt.show()


plt.hist(v,bins=np.linspace(vmin,vmax,N),ec='black')
plt.title('velocity- Histogram')
plt.xlabel('v [m/s]')
plt.show()
```

## 2. Atmosphere stability

```python
N=int(input('Give the number of data'))

tt=[]
vv=[]
ww=[]

for i in range(N):
  x=float(input('t[%.f]='%(i+1)))
  tt+=[x]
  y=float(input('v[%.f]='%(i+1)))
  vv+=[y]
  z=float(input('w[%.f]='%(i+1)))
  ww+=[z]

t=np.array(tt)
v=np.array(vv)
w=np.array(ww)

μv=np.mean(v)
σv=np.std(v)
μw=np.mean(w)
σw=np.std(w)

#Plot
plt.plot(t,v)
plt.title('v - velocity')
plt.xlabel('t [s]')
plt.ylabel('v [m/s]') 
plt.axhline(y=μv,color='red',label='mean value')
plt.legend() 
plt.show()

plt.plot(t,v)
plt.title('w - velocity')
plt.xlabel('t [s]')
plt.ylabel('w [m/s]') 
plt.axhline(y=μw,color='red',label='mean value')
plt.legend() 
plt.show()

if σw==σv:
  print('Isotropic, Conning')
elif σw<σv:
  print('Anisotropic, Fanning')
else:
  print('Anisotropic, Looping')
```
