```python
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/DAILY.csv'
df = pd.read_csv(url)
print(df)
z=np.arange(0,31,1) 
```

```python
r=scipy.stats.pearsonr(df['O3'][0:23],df['NO'][0:23])[0]

plt.figure(figsize=(5,3))
plt.xlabel('$O_{3}$ $[μg/m^{3}]$')
plt.ylabel('$NO$ $[μg/m^{3}]$')

x=[]
y=[]
for i in range(23):
  if (df['NO'][i]!=0) and (df['O3'][i]!=0):
    x+=[df['O3'][i]]
    y+=[df['NO'][i]]

plt.scatter(x, y, label='2020')

r=scipy.stats.pearsonr(x,y)[0]

m, b = np.polyfit(x,y, 1)

xx=np.linspace(60,110,40)
plt.plot(xx, m*xx + b, label='r=%.3f' %(r))
plt.legend()

plt.legend()   

plt.show()
```
