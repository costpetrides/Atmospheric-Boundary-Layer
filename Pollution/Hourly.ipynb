```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/HOURLY.csv'
df = pd.read_csv(url)
print(df)   
```

```python
plt.figure(figsize=(6,3))

plt.xlabel('Hour')
plt.ylabel('$SΟ_{2} $  $[μg/m^{3}]$')

SO2=[]
x=[]
for i in df['SO2']:
  if i==0:
    x+=[None]
  else:
    x+=[i]
    SO2+=[i]

plt.plot(df['Hour'], x, 'blue')
plt.scatter(df['Hour'], x, label='$SO_{2}$ Hourly Average')
plt.legend()

plt.xlim(df.loc[0]['Hour'] , df.loc[23]['Hour'])
plt.ylim(0,3)

plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(6,3))

plt.xlabel('Hour')
plt.ylabel('$CΟ$  $[mg/m^{3}]$')

CO=[]
x=[]
for i in df['CO']:
  if i==0:
    x+=[None]
  else:
    x+=[i]
    CO+=[i]

plt.plot(df['Hour'], x, 'purple')
plt.scatter(df['Hour'], x, color='purple' ,label='$CO$ Hourly Average')
plt.legend()

plt.xlim(df.loc[0]['Hour'] , df.loc[23]['Hour'])
plt.ylim(0,0.8)

plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(6,3))

plt.xlabel('Hour')
plt.ylabel('$NO_{x}$  $[μg/m^{3}]$')

NO=[]
x=[]
for i in df['N0']:
  if i==0:
    x+=[None]
  else:
    x+=[i]
    NO+=[i]

NO2=[]
y=[]
for i in df['NO2']:
  if i==0:
    y+=[None]
  else:
    y+=[i]
    NO2+=[i]

plt.plot(df['Hour'], x, 'green')
plt.scatter(df['Hour'], x, color='green', label='$NO$ Hourly Average')

plt.plot(df['Hour'], y, 'orange')
plt.scatter(df['Hour'], y,color='orange', label='$NO_{2}$ Hourly Average')

plt.legend()

plt.xlim(df.loc[0]['Hour'] , df.loc[23]['Hour'])
plt.ylim(0,35)

plt.xticks(rotation=90)

plt.show()
```
