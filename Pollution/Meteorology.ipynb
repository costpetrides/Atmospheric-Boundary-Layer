```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyparsing.helpers import PrecededBy
url='https://raw.githubusercontent.com/costpetrides/Atmospheric-Boundary-Layer/main/DATA/METEO.csv'
df = pd.read_csv(url)
print(df)  
xi=np.arange(0,31,1)
```

```python
plt.figure(figsize=(13,2))

plt.ylabel('Temperature  $[^{o}C]$')
plt.xlabel('Days')


plt.plot(df['DAY'], df['Tmean'], 'g', label='Temperature')
plt.scatter(df['DAY'], df['Tmean'], color='g')
plt.fill_between(df['DAY'], df['Tmin'], df['Tmax'], alpha=0.2, label='min-max')
plt.legend()

plt.xlim(df['DAY'][0] , df['DAY'][30])
plt.ylim(0,50)

plt.xticks(rotation=90)

p = np.polyfit(xi, df['Tmean'], 1)
plt.plot(df['DAY'], xi*p[0]+p[1],color='slategray')


plt.show()


print(np.mean(df['Tmean']))
print(np.std(df['Tmean']))
```

```python
plt.figure(figsize=(13,2))

plt.ylabel('RH %')
plt.xlabel('Days')



RHmean=np.zeros(31)
for i in range(31):
  RHmean[i]=(df['Rhmin'][i] + df['Rhmax'][i])/2

plt.plot(df['DAY'], RHmean, 'blue', label='Relative Humidity')
plt.scatter(df['DAY'], RHmean, color='blue')
plt.fill_between(df['DAY'], df['Rhmin'], df['Rhmax'], alpha=0.2, label='min-max')
plt.legend()


plt.legend()

plt.xlim(df['DAY'][0] , df['DAY'][30])
plt.ylim(0,100)
plt.xticks(rotation=90)


p = np.polyfit(xi, RHmean, 1)
plt.plot(df['DAY'], xi*p[0]+p[1],color='slategray')


plt.show()

print(np.mean(RHmean))
print(np.std(RHmean))
```

```python
plt.figure(figsize=(13,2))

plt.ylabel('Wind Velocity  $[Km/hr]$')
plt.xlabel('Days')

plt.plot(df['DAY'], df['Wind'], 'purple', label='Average wind Velocity')
plt.scatter(df['DAY'], df['Wind'], color='purple')
plt.legend()

plt.xlim(df['DAY'][0] , df['DAY'][30])
plt.ylim(0,40)

plt.xticks(rotation=90)

p = np.polyfit(xi, df['Wind'], 1)
plt.plot(df['DAY'], xi*p[0]+p[1],color='slategray')

plt.show()

print(np.mean(df['Wind']))
print(np.std(df['Wind']))
```
