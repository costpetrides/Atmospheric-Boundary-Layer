## Stable

```python
import numpy as np
import matplotlib.pyplot as plt

uf=0.5    # Give the friction velocity [m/s]
h=300     # Give the boundary layer depth  [m]

z=np.linspace(0,h,h)


su=2*uf*((1-(z/h)))**(3/4)
sv=2.2*uf*((1-(z/h)))**(3/4)
sw=1.73*uf*((1-(z/h)))**(3/4)

plt.plot(su,z,label='su')
plt.plot(sv,z,label='sv')
plt.plot(sw,z,label='sw')
plt.title('Stable')
plt.xlabel('Standard deviation [m/s]')
plt.ylabel('z [m]') 
plt.legend() 
plt.show()
```

## Neutral 

```python
import numpy as np
import matplotlib.pyplot as plt

uf=0.5    # Give the friction velocity [m/s]
h=300     # Give the boundary layer depth  [m]

z=np.linspace(0,h,h)


su=2.5*uf*np.exp(-1.5*z/h)
sv=1.6*uf*(1-0.5*(z/h))
sw=1.25*uf*(1-0.5*(z/h))

plt.plot(su,z,label='su')
plt.plot(sv,z,label='sv')
plt.plot(sw,z,label='sw')
plt.title('Unstable')
plt.xlabel('Standard deviation [m/s]')
plt.ylabel('z [m]') 
plt.legend() 
plt.show()
```

## Unstable

```python
import numpy as np
import matplotlib.pyplot as plt

zi=1000     # Thick Unstable boundary layer [m]
wb=30       # Give the buoyancy velocity  [m/s]

z=np.linspace(0,zi,zi)


su=0.032*wb*(1+(1-(z/zi))**6)
sv=0.032*wb
sw=0.11*wb*(z/zi)**(1/3)*(1-0.8*(z/zi))

plt.plot(su,z,label='su')
plt.axvline(x=sv,label='sv',color='orange')
plt.plot(sw,z,label='sw',color='green')
plt.title('Unstable')
plt.xlabel('Standard deviation [m/s]')
plt.ylabel('z [m]') 
plt.legend() 
plt.show()
```

Notes: Στην αστάθεια το $σ_{v}$ είναι η τυπική ταχύτητα του ανερχόμενου ‘πακέτου’ αέρα !!
